# Overview of Python programming language

Python is a high-level, interpreted programming language known for its simplicity and readability. Here's an overview of the key features and characteristics of Python:

##### 1. Readability: 
- Python's syntax is designed to be easy to read and understand, emphasizing code readability. It uses indentation and whitespace to define code blocks, making it visually appealing and reducing the need for excessive punctuation.

#### 2. Easy to Learn: 
- Python has a gentle learning curve, making it a popular choice for beginners. Its clean and straightforward syntax allows new programmers to quickly grasp the fundamentals of programming.

#### 3. Interpreted:
- Python is an interpreted language, which means that the code is executed line by line without the need for compilation. This allows for rapid development and easier debugging.

#### 4. Dynamically Typed: 
- Python is dynamically typed, meaning that you don't need to explicitly declare variable types. Variables can hold values of different types, and their types can be changed during runtime.

#### 5. Rich Standard Library: 
- Python comes with a comprehensive standard library that provides a wide range of modules and functions for various tasks, including file I/O, networking, web development, data manipulation, and more. This extensive library makes it easy to build applications without relying heavily on external libraries.

#### 6. Cross-Platform Compatibility: 
- Python is available on multiple platforms, including Windows, macOS, Linux, and many others. This allows developers to write code once and run it on different operating systems without major modifications.

####  7. Large and Active Community: 
- Python has a large and supportive community of developers worldwide. The community actively contributes to the language, creating numerous libraries, frameworks, and tools that extend Python's capabilities and address different application domains.

#### 8. Versatility: 
- Python is a versatile language used in various domains, such as web development, scientific computing, data analysis, artificial intelligence, machine learning, automation, scripting, and more. It provides the flexibility to work on different projects and adapt to changing requirements.

#### 9. Object-Oriented Programming (OOP): 
- Python supports object-oriented programming principles, allowing developers to create reusable and modular code using classes, objects, inheritance, and encapsulation.

#### 10. Integration Capabilities: 
- Python can easily integrate with other languages such as C, C++, Java, and more. This enables developers to leverage existing libraries and systems written in different languages.

#### 11. Documentation and Community Support: 
- Python has comprehensive and well-maintained documentation, making it easy to find information and learn the language. Additionally, the Python community provides forums, tutorials, and online resources, offering support to developers at all levels.

Python's versatility, simplicity, and extensive library ecosystem have contributed to its popularity and adoption in various industries and domains. It continues to be a preferred choice for both beginners and experienced programmers due to its readability, ease of use, and wide range of applications.

# Writing and running your first Python program

The print() function in Python is used to display or output information to the console or terminal. It allows you to show text, variables, or the results of expressions during program execution. The print() function is a commonly used tool for debugging, displaying program outputs, and providing information to users.

Syntax:
The basic syntax of the print() function is as follows
______________________________________________________________________________________________

--> print(value1, value2, ..., sep=' ', end='\n', file=sys.stdout, flush=False)

______________________________________________________________________________________________


- value1, value2, ...: These are the values or expressions that you want to display. You can provide multiple values separated by commas.
- sep=' ': This is an optional parameter that specifies the separator between the values. By default, it is set to a space character (' ').
- end='\n': This is an optional parameter that specifies the string that will be appended after all the values are printed. By default, it is set to a newline character ('\n'), which moves the cursor to the next line after printing.
- file=sys.stdout: This is an optional parameter that specifies the output file. By default, it is set to the standard output file (sys.stdout), which represents the console or terminal.
- flush=False: This is an optional parameter that specifies whether the output should be flushed immediately. By default, it is set to False, meaning the output is buffered and may not be displayed immediately.


````python
print('Hello World')
````

````python
#Printing variables:

name = "Alice"
age = 25
print("Name:", name, "Age:", age)
````

# sep parameter:

- The sep parameter specifies the separator between the values passed to the print() function.
- By default, the separator is a space character (' ').
- You can customize the separator by assigning a different string to the sep parameter.

````python
# Specifying a custom separator:
print("apple", "banana", "orange", sep=', ')
````

# end parameter:

- The end parameter specifies the string to be appended after all the values are printed.
- By default, the ending character is a newline character ('\n'), which moves the cursor to the next line.
- You can modify the ending character by assigning a different string to the end parameter.

````python
# Changing the ending character

print("Hello", end='!')
print("Welcome")
````

# file parameter:

- The file parameter allows you to specify the output file where the values will be printed.
- By default, it is set to sys.stdout, which represents the console or terminal.
- You can redirect the output to a file by assigning a file object to the file parameter.

````python
with open("output.txt", "w") as f:
    print("Hello, World!", file=f)
#create new file output.txt in directory and print hello,World! in that.
````

# flush parameter:

- The flush parameter determines whether the output should be flushed immediately.
- By default, it is set to False, meaning the output is buffered and may not be displayed immediately.
- Setting flush to True forces the immediate display of the output.

````python
print("Processing...", flush=True)
````

# input()
The `input()` function in Python is used to prompt the user for input. It allows you to interact with the user by displaying a prompt message and waiting for the user to enter a value from the keyboard.

The syntax of the `input()` function is as follows:

```python
variable = input(prompt)
```

- `prompt` is an optional parameter that specifies the message to be displayed to the user. It serves as a prompt to indicate what kind of input is expected.

The `input()` function reads the user's input as a string. If you need to interpret the input as a different data type (e.g., integer, float), you can explicitly convert the input using appropriate typecasting functions like `int()` or `float()`.



````python

name = input("Enter your name: ")
print("Hello,", name)

age = input("Enter your age: ")
age = int(age)  # Converting the input to an integer

next_year_age = age + 1
print("Next year, you will be", next_year_age, "years old.")

````


In this example, the `input()` function is used to prompt the user for their name and age. The inputs are stored in variables (`name` and `age`), and then the program performs some calculations based on the user's age.

When the program executes, it displays the prompt message, waits for the user to enter the input, and then proceeds with the further execution.

Note: It's important to handle user input carefully, especially when using it in calculations or operations. Consider handling exceptions or validating the input to ensure it meets your program's requirements.
